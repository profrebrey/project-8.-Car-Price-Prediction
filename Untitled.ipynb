### name

Это поле содержит информацию, которая в основном содержится в других признаках, однако есть аттрибуты, которые дают дополнительную информацию. Выделим ключевые слова и разделим их по основным группам. С учетом того, что встречаются вариции, выделим группы вручную.

```python
import nltk
nltk.download('punkt')
from nltk import word_tokenize
nltk.download('stopwords')
from nltk.corpus import stopwords
english_stopwords = stopwords.words("english")
import string
spec_chars = string.punctuation + '«»—…’‘”“©'

# Зададим функцию для удаления спец символов
def remove_chars_from_text(text, chars):
    return "".join([ch for ch in text if ch not in chars])
```

Определим функцию токенизации

```python
def text_tokenizer(text):
    text_ = text.strip()
    text_ = text_.lower()
    text_ = remove_chars_from_text(text_, spec_chars)
    text_ = remove_chars_from_text(text_, string.digits)

    text_ = text_.replace('\n',' ').replace('\t',' ')
    tokens = word_tokenize(text_)
    token_text = nltk.Text(tokens)
    filtered_token_text = [w for w in token_text if not w in english_stopwords_ and len(w) > 2]
    f_tokens = []
    for _t in filtered_token_text:
        for t_name, kw in name_keywords.items():
            if _t in kw:
                f_tokens.append(t_name)
    
    return f_tokens
```

Выясним, какие токены встречаются и выделим группы слов, которые будем использовать как категориальные признаки.

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import RobustScaler
def preproc_data(df_input, numerical_features, cf_wo_dummy):
    '''includes several functions to pre-process the predictor data.'''

    df_output = df_input.copy()
    # Убираем не нужные для модели признаки
    df_output.drop(['description', 'sell_id', 'vehicleConfiguration',
                   'name', 'Владение', 'modelDate'], axis=1, inplace=True)

    # Заполняем пропуски
    for column in numerical_features:
        df_output[column].fillna(df_output[column].median(), inplace=True)
    
    # Стандартизацию используем усточивую к выбросам, потому что судя по распределению числовых данных, они встречаются
    scaler = RobustScaler() 
    for column in numerical_features:
        df_output[column] = scaler.fit_transform(df_output[[column]])[:, 0]

    # Label Encoding для категориальных признаков
    for column in cf_wo_dummy:
        df_output[column] = df_output[column].astype('category').cat.codes

    # генерируем dummies
    df_output = pd.get_dummies(
        df_output, columns=cf_wo_dummy, dummy_na=False)
        
    return df_output
```

### descriptions

Объявления о продаже довольно похожи, поэтому стоит попробовать выделить основные тематики объявлений - ключевые слова, подходящие к каждому объявлению (темы объявлений).
Применим LDA для генерации тем.

```python
descriptions = data.description.to_list()
```

```python
stop = set(stopwords.words('russian'))
exclude = set(string.punctuation) 
lemma = WordNetLemmatizer() 

# Метод для очистки текстов, лематизации и нормализации
def clean(doc):
    stop_free = " ".join([i for i in doc.lower().split() if i not in stop])
    punc_free = ''.join(ch for ch in stop_free if ch not in exclude)  
    normalized = " ".join(lemma.lemmatize(word) for word in punc_free.split())
    normalized = re.sub(r'(?:[\W\d_])+', ' ', normalized)
    return normalized

clean_descriptions = [clean(doc).split() for doc in descriptions]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation

tf_idf_vectorizer = TfidfVectorizer(tokenizer=lambda doc: doc, lowercase=False)
cv_vectorizer = CountVectorizer(tokenizer=lambda doc: doc, lowercase=False)
```

```python
tf_idf_arr = tf_idf_vectorizer.fit_transform(clean_descriptions)
cv_arr = cv_vectorizer.fit_transform(clean_descriptions)
```

Составляем словари в векторизованном виде

```python
vocab_tf_idf = tf_idf_vectorizer.get_feature_names()
vocab_cv = cv_vectorizer.get_feature_names()
```

Генерируем 100 возможных тем, которые будут описывать тексты

```python
TOPIC_NUM = 100
lda_model = LatentDirichletAllocation(n_components = TOPIC_NUM, max_iter = 20, random_state = RANDOM_SEED)
X_topics = lda_model.fit_transform(tf_idf_arr)
topic_words = lda_model.components_
```

Каждая тема будет состоять из 5 самых популярных слов

```python
n_top_words = 5

topics = {}

for i, topic_dist in enumerate(topic_words):
    sorted_topic_dist = np.argsort(topic_dist)
    topic_words = np.array(vocab_tf_idf)[sorted_topic_dist]
    topic_words = topic_words[:-n_top_words:-1]
    topics[i+1] = topic_words
```

Собираем темы для каждого текста и генерируем дамми признаки на основе ключевых слов каждой темы

```python
all_topics = []

doc_topic = lda_model.transform(tf_idf_arr)  
for n in range(doc_topic.shape[0]):
    topic_doc = doc_topic[n].argmax()
    _t = [t for t in list(topics[topic_doc+1]) if len(t) > 2]
    all_topics.append(",".join(_t))
```

```python
df_topics = pd.DataFrame(np.array(all_topics))
df_topics = df_topics[0].str.get_dummies(',')
df_topics
```

## NLP + Multiple Inputs

Построим модель нейронной  на основе описаний

```python
MAX_WORDS = 100000
MAX_SEQUENCE_LENGTH = 256
```

```python
text_train = data.description.iloc[X_train.index]
text_test = data.description.iloc[X_test.index]
text_sub = data.description.iloc[X_sub.index]
```

```python
nltk.download('wordnet')
```

Очистим тексты от стоп-слов и проиведем лемматизацию

```python
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
from tensorflow.keras.preprocessing.text import Tokenizer
from tqdm import tqdm
from keras.preprocessing.text import text_to_word_sequence

lemmatizer = WordNetLemmatizer()
tokenize = Tokenizer(num_words=MAX_WORDS)
stop_words = set(stopwords.words("russian"))

def preprocces(X):
  X_proccess = []

  for x in tqdm(X):
    
    x = x.lower()
    x = text_to_word_sequence(x)
    tokenize.fit_on_texts(x)
    x = [word for word in x if word.isalnum()]
    x = [lemmatizer.lemmatize(w) for w in x]
    x = [word for word in x if not word in stop_words]
    X_proccess.append(' '.join(x))
  return X_proccess


text_train_proc = preprocces(text_train)
text_test_proc = preprocces(text_test)
text_sub_proc = preprocces(text_sub)
```

Производим вектоизацию текста

```python
text_train_sequences = sequence.pad_sequences(tokenize.texts_to_sequences(text_train_proc), maxlen=MAX_SEQUENCE_LENGTH)
text_test_sequences = sequence.pad_sequences(tokenize.texts_to_sequences(text_test_proc), maxlen=MAX_SEQUENCE_LENGTH)
text_sub_sequences = sequence.pad_sequences(tokenize.texts_to_sequences(text_sub_proc), maxlen=MAX_SEQUENCE_LENGTH)

print(text_train_sequences.shape, text_test_sequences.shape, text_sub_sequences.shape, )
```

Строим сеть для обработки текста. По сравнению с бейзлайном добавим полносвязный слой.

```python
model_nlp = Sequential()
model_nlp.add(L.Input(shape=MAX_SEQUENCE_LENGTH))
model_nlp.add(L.Embedding(len(tokenize.word_index)+1, MAX_SEQUENCE_LENGTH,))
model_nlp.add(L.LSTM(256, return_sequences=True))
model_nlp.add(L.Dropout(0.5))
model_nlp.add(L.Dense(64, activation="sigmoid"))
model_nlp.add(L.LSTM(128,))
model_nlp.add(L.Dropout(0.25))
model_nlp.add(L.Dense(64, activation="relu"))
model_nlp.add(L.Dropout(0.25))
```

И так же используем сеть, которая была построена для табличных данных

```python
model_mlp = Sequential()
model_mlp.add(L.Dense(512, input_dim=X_train.shape[1], activation="sigmoid"))
model_mlp.add(L.Dropout(0.5))
model_mlp.add(L.Dense(256, activation="relu"))
model_mlp.add(L.Dense(256, activation="relu")) #добавим еще один полносвязный слой
model_mlp.add(L.Dropout(0.5))
```

Объединим две сети в одну

```python
combinedInput = L.concatenate([model_nlp.output, model_mlp.output])
head = L.Dense(64, activation="relu")(combinedInput)
head = L.Dense(1, activation="linear")(head)

model = Model(inputs=[model_nlp.input, model_mlp.input], outputs=head)
```

Обучаем

```python
optimizer = tf.keras.optimizers.Adam(0.01)
model.compile(loss='MAPE',optimizer=optimizer, metrics=['MAPE'])
```

```python
checkpoint = ModelCheckpoint('../working/best_model.hdf5', monitor='val_MAPE',
                             save_best_only=True, verbose=0, mode='min')
earlystop = EarlyStopping(monitor='val_MAPE', patience=10, restore_best_weights=True,)
callbacks_list = [checkpoint, earlystop]
```

```python
history = model.fit([text_train_sequences, X_train], y_train,
                    batch_size=512,
                    epochs=350,
                    validation_data=([text_test_sequences, X_test], y_test),
                    callbacks=callbacks_list
                   )
```

```python
plt.title('Loss')
plt.plot(history.history['MAPE'], label='train')
plt.plot(history.history['val_MAPE'], label='test')
plt.show();
```

```python
AUGMENTATIONS = albumentations.Compose([
    albumentations.HorizontalFlip(p=0.5),
    albumentations.Rotate(limit=30, interpolation=1, border_mode=4,
                          value=None, mask_value=None, always_apply=False, p=0.5),
    albumentations.OneOf([
        albumentations.CenterCrop(height=224, width=200),
        albumentations.CenterCrop(height=200, width=224),
    ], p=0.5),
    albumentations.OneOf([
        albumentations.RandomBrightnessContrast(
            brightness_limit=0.3, contrast_limit=0.3),
        albumentations.RandomBrightnessContrast(
            brightness_limit=0.1, contrast_limit=0.1)
    ], p=0.5),
    albumentations.GaussianBlur(p=0.05),
    albumentations.HueSaturationValue(p=0.5),
    albumentations.RGBShift(p=0.5),
    albumentations.FancyPCA(alpha=0.1, always_apply=False, p=0.5),
    albumentations.Resize(240, 320)
])
```

```python
def make_augmentations(images):
  print('применение аугментаций', end = '')
  augmented_images = np.empty(images.shape)
  for i in range(images.shape[0]):
    if i % 200 == 0:
      print('.', end = '')
    augment_dict = AUGMENTATIONS(image = images[i])
    augmented_image = augment_dict['image']
    augmented_images[i] = augmented_image
  print('')
  return augmented_images
```

```python
data_NLP = data.copy()
tokenize = Tokenizer(num_words=MAX_WORDS)
tokenize.fit_on_texts(data_NLP['description'])
```

```python
def process_image(image):
    return AUGMENTATIONS(image=image.numpy())['image']


def tokenize_(descriptions):
    return sequence.pad_sequences(tokenize.texts_to_sequences(descriptions), maxlen=MAX_SEQUENCE_LENGTH)


def tokenize_text(text):
    return tokenize_([text.numpy().decode('utf-8')])[0]


def tf_process_train_dataset_element(image, table_data, text, price):
    im_shape = image.shape
    [image, ] = tf.py_function(process_image, [image], [tf.uint8])
    image.set_shape(im_shape)
    [text, ] = tf.py_function(tokenize_text, [text], [tf.int32])
    return (image, table_data, text), price


def tf_process_val_dataset_element(image, table_data, text, price):
    [text, ] = tf.py_function(tokenize_text, [text], [tf.int32])
    return (image, table_data, text), price


train_dataset = tf.data.Dataset.from_tensor_slices((
    images_train, X_train, data.description.iloc[X_train.index], y_train
)).map(tf_process_train_dataset_element)

test_dataset = tf.data.Dataset.from_tensor_slices((
    images_test, X_test, data.description.iloc[X_test.index], y_test
)).map(tf_process_val_dataset_element)

y_sub = np.zeros(len(X_sub))
sub_dataset = tf.data.Dataset.from_tensor_slices((
    images_sub, X_sub, data.description.iloc[X_sub.index], y_sub
)).map(tf_process_val_dataset_element)
```

```python
efficientnet_model = tf.keras.applications.efficientnet.EfficientNetB3(weights='imagenet',
                                                                       include_top=False, input_shape=(size[1], size[0], 3))
```

```python
efficientnet_model.trainable = True
fine_tune_at = len(efficientnet_model.layers)//2

# Keep all other layers frozen
for layer in efficientnet_model.layers[:fine_tune_at]:
    layer.trainable =  False
print(len(efficientnet_model.layers))
```

```python
efficientnet_output = L.GlobalAveragePooling2D()(efficientnet_model.output)
```

```python
tabular_model = Sequential([
    L.Input(shape = X.shape[1]),
    L.Dense(512, activation = 'sigmoid'),  # изменим функцию активации
    L.LayerNormalization(),
    L.Dropout(0.5),
    L.Dense(256, activation = 'relu'),#добавили еще один полносвязный слой 
    L.Dense(256, activation = 'relu'),
    L.Dropout(0.5),
    ])
```

```python
nlp_model = Sequential([
    L.Input(shape=MAX_SEQUENCE_LENGTH),
    L.Embedding(len(tokenize.word_index)+1, MAX_SEQUENCE_LENGTH,),
    L.LayerNormalization(),
    L.LSTM(256, return_sequences=True),
    L.Dropout(0.5),
    L.Dense(128, activation = 'sigmoid'), 
    L.Dropout(0.5),
    L.LSTM(64),
    L.Dropout(0.25),
    L.Dense(64),
    L.Dropout(0.25)
    ])
```

```python
combinedInput = L.concatenate(
    [efficientnet_output, tabular_model.output, nlp_model.output])
head = L.Dense(256, activation="relu")(combinedInput)
head = L.Dense(1,)(head)

model = Model(inputs=[efficientnet_model.input,
                      tabular_model.input, nlp_model.input], outputs=head)
# model.summary()
```

```python
optimizer = tf.keras.optimizers.Adam(0.005)
model.compile(loss='MAPE',optimizer=optimizer, metrics=['MAPE'])
```
